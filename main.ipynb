# Dataset


Our analysis will be done on the QQQ's. Lets do a really really simple baseline modeL


```python
qqq_df_original = pd.read_csv('data/historical/etfs/QQQ.csv')
next_row_values = qqq_df_original.shift(-1)["Close"] 
qqq_df_original["Next day closing"] = next_row_values
# Drop last row
last_row = len(qqq_df_original)
qqq_df = qqq_df_original.drop(qqq_df_original.index[last_row-1]) 
qqq_df
```

```python
qqq_df['Date'] = pd.to_datetime(qqq_df['Date'])
```

```python
# Extract year, month, and day
qqq_df['Year'] = qqq_df['Date'].dt.year
qqq_df['Month'] = qqq_df['Date'].dt.month
qqq_df['Day'] = qqq_df['Date'].dt.day
qqq_df
```

```python
qqq_df.describe()
```

Lets add a proper target column. Lets target the price for the next day


For the first part lets split the data from train and test, since this is 

```python
from sklearn.model_selection import TimeSeriesSplit

target = 'Next day closing'
test_size = 0.2
#Calculate Feature and target list
X = qqq_df.drop(columns=[target])
y = qqq_df[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

n_splits = 5 
tscv = TimeSeriesSplit(n_splits=n_splits)
```

Lets do some EDA

```python
X_train.info()
```

```python
y_train.info()
```

Now lets do a base line model. I will be using Dummy Regressor

```python
dummyRegressor = DummyRegressor(strategy="mean")
dummyRegressor.fit(X_train,y_train)
```

```python
y_pred = dummyRegressor.predict(X_test)

# Compare raw predictions against actual values
comparison_df = pd.DataFrame({'Date': X_test['Date'], 'Actual': y_test, 'Predicted': y_pred})

# Print or display the comparison DataFrame
print(comparison_df)
```

```python
r2 = cross_val_score(dummyRegressor, X_train, y_train, cv=tscv, scoring = 'r2', n_jobs =1).mean()
r2
```

```python
comparison_df = comparison_df.sort_values(by='Date')

# Plot the comparison
plt.figure(figsize=(12, 6))
plt.plot(comparison_df['Date'], comparison_df['Actual'], label='Actual', marker='o',markersize=2)
plt.plot(comparison_df['Date'], comparison_df['Predicted'], label='Predicted', marker='o',markersize=2)

plt.title('Actual vs. Predicted Next Day Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
